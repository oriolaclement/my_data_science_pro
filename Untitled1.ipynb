```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='winter_r')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ['Apples', 'Oranges', 'Bananas', 'Melons']
myexplode = [0.2, 0.3, 0.4, 0.1]
mycolors = ['black', 'green', 'cyan', 'hotpink']

plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode,shadow = True, colors = mycolors)
plt.legend(title= 'MY_FRUITS')
plt.show() 
```
