```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
font1 ={'family':'serif', 'color':'black', 'size':20}
font2 ={'family':'serif', 'color': 'blue', 'size':20}
plt.title('class_chart', fontdict = font1, loc = 'right')
plt.xlabel('number of students',fontdict=font1)
plt.ylabel('total_score', fontdict=font2)
plt.grid(axis='x')
plt.plot(xpoints, ypoints,'o:b')
plt.show()
```
